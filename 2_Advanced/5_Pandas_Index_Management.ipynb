```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

## Problems

```python
# 1_Problem Statement:
# Set the job_posted_date column as the index of the DataFrame and display the first 5 rows.

df_us = df[df['job_country'] == 'United States']

df_us.set_index('job_posted_date', inplace=True)
df_us.head(5)
```

```python
# 2_Problem Statement:
# Reset the index of the DataFrame from the last problem (where job_posted_date column is the index) to the default integer index 
# and drop the existing index. Display the first 5 rows.

df_us.reset_index(inplace=True)
df_us.head(5)
```

```python
# 3_Problem Statement:
# Create a pivot table to calculate the median salary_year_avg for each company_name and sort the index in descending order. 
# Display the first 5 rows.

pivot_table_us = df_us.pivot_table(values='salary_year_avg', index='company_name', aggfunc='median')
pivot_table_us.sort_index(inplace=True, ascending=False)
pivot_table_us.head(5)
```

```python
# 4_Problem Statement:
# Create a pivot table to calculate the median salary_year_avg for each job_title_short, 
# reset the index of the resulting pivot table, then set the index to salary_year_avg. Display the first 5 rows.

pivot_tab = df_us.pivot_table(values='salary_year_avg', index='job_title_short', aggfunc='median')
pivot_tab.reset_index(inplace=True)
pivot_tab.set_index('salary_year_avg', inplace=True)
pivot_tab.head(5)
```
